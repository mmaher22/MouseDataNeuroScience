```python
import os
import math
import time
import random
import scipy.io
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
most_images_data = scipy.io.loadmat('data/natimg2800_M170714_MP032_2017-09-14.mat')
images_categories = scipy.io.loadmat('data/stimuli_class_assignment_confident.mat')
resp = most_images_data['stim'][0]['resp'][0]   
istim = most_images_data['stim'][0]['istim'][0] 
cats = images_categories['class_assignment'][0] # cateogries of stimuli
labels_str = np.array([z[0] for z in images_categories['class_names'][0]])
```

```python
X_data = pd.DataFrame(data = resp)
y_data = pd.DataFrame(data = istim)
```

```python
def id_to_label(x):
    return labels_str[cats[x-1]] if x < 2801 else 'grey'
y_data[0] = y_data[0].apply(id_to_label)
```

```python
classes, counts = y_data[0].value_counts().index, y_data[0].value_counts().values
for idd, c in zip(classes, counts):
    print(f'{idd}    \t', c)
```

```python
plt.figure()
plt.bar(classes, counts, width=0.5)
plt.title('Proportions of each class', fontsize = 16)
plt.xlabel('Class Label', fontsize = 16)
plt.ylabel('Frequency', fontsize = 16)
plt.xticks(rotation='vertical')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=22)
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()
```

```python
clf = RandomForestClassifier(n_estimators=200, max_depth=50, random_state=22)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, y_pred))
```

**I will now try to visualize different non linear feature selection methods by taking the first 2 components and plotting them. I will also try with 3 components and 3D plots**

```python
from sklearn import manifold
n_neighbors = 10
iso_data = manifold.Isomap(n_neighbors, n_components=2).fit_transform(X_data)
```

```python
iso_data.shape
```

```python
palette = sns.color_palette("bright", 13)
plt.figure(figsize=(20,15))
sns.scatterplot(iso_data[:,0], iso_data[:,1], legend = 'full', palette = palette, hue = y_data.values[:,0])
```

```python
X_train, X_test, y_train, y_test = train_test_split(iso_data, y_data, test_size=0.2, random_state=22)
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()

clf = RandomForestClassifier(n_estimators=200, max_depth=50, random_state=22)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, y_pred))
```

**I will try ISOmaps again but with more components**

```python
n_neighbors = 10
iso_data = manifold.Isomap(n_neighbors, n_components=100).fit_transform(X_data)
```

```python
palette = sns.color_palette("bright", 13)
plt.figure(figsize=(20,15))
sns.scatterplot(iso_data[:,0], iso_data[:,1], legend = 'full', palette = palette, hue = y_data.values[:,0])
```

```python
X_train, X_test, y_train, y_test = train_test_split(iso_data, y_data, test_size=0.2, random_state=22)
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()

clf = RandomForestClassifier(n_estimators=200, max_depth=50, random_state=22)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, y_pred))
```

**I will now try kernel PCA**

```python
from sklearn.decomposition import KernelPCA
kpca_data = KernelPCA(n_components=100, kernel='rbf').fit_transform(X_data)
```

```python
palette = sns.color_palette("bright", 13)
plt.figure(figsize=(20,15))
sns.scatterplot(kpca_data[:,0], kpca_data[:,1], legend = 'full', palette = palette, hue = y_data.values[:,0])
```

```python
X_train, X_test, y_train, y_test = train_test_split(kpca_data, y_data, test_size=0.2, random_state=22)
y_train = y_train.values.ravel()
y_test = y_test.values.ravel()

clf = RandomForestClassifier(n_estimators=200, max_depth=50, random_state=22)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, y_pred))
```
